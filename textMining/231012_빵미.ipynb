```python
import pandas as pd
```

```python
c_data_df = pd.read_csv('커피빈매장정보.csv', encoding='euc-kr')
c_data_df
```

```python
h_data_df = pd.read_csv('할리스매장정보.csv', encoding='utf-8')
h_data_df
```

```python
c_area_df = c_data_df['address'].str[:2]
c_area_df
c_area_df = c_area_df.replace('경상', '경남')
```

```python
type(c_area_df.value_counts())
```

```python
c_vc = c_area_df.value_counts()
c_vc
```

```python
h_area_df = h_data_df['area'].str[:2]
h_area_df
```

```python
h_vc = h_area_df.value_counts()
h_vc
```

```python
df1 = pd.DataFrame({'hollys': h_vc})
df2 = pd.DataFrame({'coffee bean': c_vc})
result_df = pd.concat([df1, df2], axis=1)
result_df
```

```python
df = result_df.fillna(0)
df
```

```python
df = df.astype(int)
df
```

```python
df.reset_index(level=0, inplace=True)
df
```

```python
df = df.rename(columns={'지역': 'area'})
df
```

```python
import matplotlib.pyplot as plt
```

```python
ax = df.plot(x='area', y=['hollys', 'coffee bean'], kind='bar')
plt.title('할리스 커피빈 지역별 비교')
plt.xlabel('area')
plt.ylabel('hollys, coffeebean')
plt.show()
```
