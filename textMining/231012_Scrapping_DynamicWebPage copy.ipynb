## 커피빈 - 동적 웹페이지

## dataframe화, CSV 변경

```python
import pandas as pd


h_data = pd.read_csv('할리스매장정보.csv')

h_data
```

```python
h_address = h_data['area'].str[:2]
h_address.unique()
h_area_cnt = h_address.value_counts()
h_area_cnt # 지역별 개수(series)
```

```python
b_data = pd.read_csv('231011_커피빈매장정보.csv')

b_data 
```

```python
b_address = b_data['address'].str[:2]
b_address.replace('경상', '경남', inplace=True)
# b_address.unique()
b_area_cnt = b_address.value_counts()
b_area_cnt # 지역별 개수(series)
```

```python
total_cafe_df = pd.DataFrame(
    {'할리스' : h_area_cnt,
     '커피빈' : b_area_cnt
     }
)
total_cafe_df.sort_values(by='할리스', ascending=False, inplace=True)
total_cafe_df
```
