```python
import matplotlib.pyplot as plt

with open('data/seoul-2022.csv', 'r') as file: 
    data = csv.reader(file)
    highTemplist = list()
    next(data)
    for row in data:
#         print(raw[-1])
        if len(row)>1 and row[-1] != '':
            highTemplist.append(float(row[-1]))
            
plt.figure(figsize=(10, 3)) #int 기준인듯.
plt.plot(highTemplist, 'r')
```

```python
import csv

# 매년 8월의 최고기온을 시각화
with open('data/seoul-1970to2023.csv', 'r') as file:
    data = csv.reader(file)
    highTempList = list()
    next(data)
    for row in data:
        # 길이가 비지 않았으면서, 값이 ''비지 않은 애.
        if len(row) > 1 and row[-1] != '':
            date = row[0].split('-')
            month = date[1]
            if month == '08':
#                 print(row[-1])
                highTempList.append(float(row[-1]))
    plt.figure(figsize=(10,3))
    plt.plot(highTemplist, 'r')
```

```python
with open('data/seoul-1970to2023.csv', 'r') as file:
    data = csv.reader(file)
    next(data)
    myLowTempList = []
    myHighTempList = []
    years = []
    
    for row in data:
        if len(row)>1 and row[-1] != '':
#             print(row[0].split('-'))
            date = row[0].split('-')
            month = date[1]
            day = date[2]
            if month == '12' and day == '08':
#                 print(row)
                myHighTempList.append(float(row[-1]))
                myLowTempList.append(float(row[-2]))
                years.append(date[0].strip())
                
        
#     print(len(myLowTempList))
#     print(len(myHighTempList))

    plt.figure(figsize=(15,3))
    plt.title('Low&High Temperature on my Birthday')

    #x축의 각도를 90도로 틀어서 겹치지 않도록.
    plt.xticks(rotation=90)
    #x축, y축에 들어갈 데이터를 지정하는 것도 가능.
    plt.plot(years, myHighTempList,'r')
    plt.plot(years, myLowTempList,'b')
```
