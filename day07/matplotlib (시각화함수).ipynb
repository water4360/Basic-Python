```python
import matplotlib.pyplot as plt
```

```python
plt.title('plot test')
plt.plot([10,20,30,40])
# plt.show() 주피터에서는 생략 가능.
```

```python
# X축, Y축
plt.plot([10,20,30,40], [1,2,3,4])
```

```python
# X축, Y축 크로스
#단 key없이 color 적는 경우 순서 지킬 것.
#linestyle은 solid(-), dashed(--), dotted(:), dash-dot(-.)
#컬러명도 축약 가능 r, g, b 등
plt.plot([10,20,30,40], label='asc',  color='pink', linestyle='--')
plt.plot([40,30,20,10], 'g', label='desc', ls=':')
# plt.legend()
plt.legend(loc=6)
```

```python
plt.plot([10,20,30,40],'g.', ls='-.')
plt.plot([40,30,20,10],'b^')
```

```python
import matplotlib.pyplot as plt

# 히스토그램(빈도수)
xList = [1, 10, 20, 30, 40]
yList = [5, 20, 14, 50, 36]

plt.bar(xList, yList)
plt.barh(xList, yList)
```

```python
data = [10, 1,20,20,5, 30,20,30,20,10,10,1,30]
plt.hist(data, bins=10) #bins는 30칸으로 분할해서 보는 것
```

```python
# 주사위를 10번 던져 1~6 숫자의 빈도수 시각화
import random
numbers = []
for i in range(10000):
    numbers.append(random.randint(1,6))
print(numbers)

plt.hist(numbers, bins=6)
```

```python
numbers = []
for i in range(11):
    numbers.append(random.randint(1, 100))
print(sorted(numbers))

plt.boxplot(numbers)
```
