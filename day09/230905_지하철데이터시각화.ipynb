# 유임승차, 하차 / 무임승차, 하차 승객수가 가장 많은 역과 승객수 조회

```python
import csv
import matplotlib.pylab as plt

#UnicodeDecodeError 발생하는 경우,
#문서의 인코딩을 UTF-8로 변경할 것.
with open('data/202308_subway.csv', 'r', encoding='utf-8') as file:
    data = csv.reader(file)
    next(data)
    #역명, 인원수
    #maxInfo[0] : 유임승차 승객수가 가장 많은 역
    #       [1] : 해당 역의 승객수
    maxInfo = ['', -1]
    offInfo = ['', -1]
    
    freeMaxInfo = ['', -1]
    freeOffInfo = ['', -1]
    
    totalInfo = ['', -1, '', -1, '', -1, '', -1]
    
    for row in data:
        #값 비교시, 데이터 타입(문자)에 주의할 것.
        row[4] = int(row[4].replace(',',''))
        row[5] = int(row[5].replace(',',''))
        row[6] = int(row[6].replace(',',''))
        row[7] = int(row[7].replace(',',''))
        
        if row[4] > maxInfo[1]:
            maxInfo[0] = row[3]
            maxInfo[1] = row[4]
    
        if row[5] > offInfo[1]:
            offInfo[0] = row[3]
            offInfo[1] = row[5]
        
        if row[6] > freeMaxInfo[1]:
            freeMaxInfo[0] = row[3]
            freeMaxInfo[1] = row[6]
        
        if row[7] > freeOffInfo[1]:
            freeOffInfo[0] = row[3]
            freeOffInfo[1] = row[7]
        
            
    print(f'[최다 유임승차] {maxInfo[0]}역, 인원 {maxInfo[1]}명')
    print(f'[최다 유임하차] {offInfo[0]}역, 인원 {offInfo[1]}명')
    print(f'[최다 무임승차] {freeMaxInfo[0]}역, 인원 {freeMaxInfo[1]}명')
    print(f'[최다 무임하차] {freeOffInfo[0]}역, 인원 {freeOffInfo[1]}명')
```

```python
with open('data/202308_subway.csv', 'r', encoding='utf-8') as file:
    data = csv.reader(file)
    next(data)
    
    #역 따로 승객수 따로 저장해보려고 함
    stations = ['']*4
    max_persons = [0]*4
#     print(stations, max_persons)
    for row in data:
        for i in range(4):
            person = int(row[4+i].replace(',',''))
            if person > max_persons[i]:
                stations[i]= row[3]
                max_persons[i]= person
#     print(stations)
#     print(max_persons)
    titles = ['유임승차','유임하차','무임승차','무임하차']
    for i in range(len(stations)):
        print(f'[최다 {titles[i]}]{stations[i]}, {max_persons[i]}')
```

# 출근시간(07시)에 가장 많이 승차하는 역 검색

```python
with open('data/202308_timeline.csv', 'r', encoding='utf-8') as file:
    data = csv.reader(file)
    next(data)
    next(data)
    
    #알고싶은 정보 초기화
    station = ''
    max_persons = 0
    
    for row in data:
        person = int(row[10].replace(',',''))
        if person > max_persons:
            station = f'{row[3]}({row[1]})'
            max_persons = person
            
    print(f'{station} : {max_persons}명')
```

# 출근시간대(07~09시)에 가장 많이 승차하는 역

```python
def to_int(data):
    return int(data.replace(',',''))

with open('data/202308_timeline.csv', 'r', encoding='utf-8') as file:
    data = csv.reader(file)
    next(data)
    next(data)
    
    #알고싶은 정보 초기화
    station = ''
    max_persons = 0
    
    for row in data:
        person = 0
        #인덱스로 쓸때. 10, 12
        for i in range(2):
            person += int(row[10+(i*2)].replace(',',''))
        if person > max_persons:
            station = row[3]
            max_persons = person
            
    print(f'{station} : {max_persons}명')
```

# 검색할 시간대 입력(24시) : 9
## n시 대에 가장 많은 승차인원을 가진 역과 인원수 출력

```python
import csv
with open('data/202308_timeline.csv', 'r', encoding='utf-8') as file:
    data = csv.reader(file)
    next(data)
    next(data)
    
    inputTime = int(input('검색할 시간대를 입력하세요(0~24) : '))
    #  [n+i]번째 데이터
    
    station =''
    max_persons = 0
    index = inputTime * 2 - 4
    
    for row in data:
        row[index] = int(row[index].replace(',',''))
        if max_persons < row[index]:
            max_persons = row[index]
            station = f'{row[3]}({row[1]})'
    print(f'{inputTime}시 대 : {station}[{max_persons}명]')
```
