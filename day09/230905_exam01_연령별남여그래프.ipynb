```python
import csv
import matplotlib.pyplot as plt
```

```python
with open('data/202308_gender.csv', 'r') as file:
    data = csv.reader(file)
    header=next(data)
    print(header[3], header[3+100])
    print(header[106], header[106+100])
    town = '성복'
    
    for row in data:
        if town in row[0]:
            male = list()
            female = list()
            for i in range(0,101): #0세부터 100세까지.
                male.append(int(row[i+3].replace(',',''))) #i[3]부터 0세니까.
                female.append(int(row[i+106].replace(',','')))
    print(male)
    print(female)
```

# 남성인구수를 음수로 변환하여 항아리형 그래프로.

```python
for i in range(len(male)):
    male[i] = -male[i]
print(male)
```

```python
plt.barh(range(101), male)
plt.barh(range(101), female)
```

# list comprehension 이용하기

```python
with open('data/202308_gender.csv', 'r') as file:
    data = csv.reader(file)
    header=next(data)
    town = '성복'
    
    for row in data:
        if town in row[0]:
#             print(row[3:3+100+1])
            male = [int(i.replace(',','')) for i in row[3:104]]
            female = [int(i.replace(',','')) for i in row[106:]]
    
    male = [-i for i in male]
    print(male)
    print(female)
```

```python
#한글깨지지 않도록 font 변경해주기
#rc = resource
plt.rc('font', family='gulim')
plt.title(f'{town}동 연령별 남/여 그래프')

plt.barh(range(101), male, label='남성')
plt.barh(range(101), female, label='여성')

plt.legend() #범례 추가
plt.xlabel('population')
plt.ylabel('age')
```
