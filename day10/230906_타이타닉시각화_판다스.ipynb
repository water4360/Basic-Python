```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# raw_data = pd.read_excel('data/titanic.xls')
row_data = pd.read_csv('data/titanic.csv', sep='\t')
row_data
```

```python
row_data.head(10)
# row_data.tail(10)
```

```python
row_data.info()
```

```python
row_data.describe() #간단한 통계 조회가능
```

```python
# plt.plot(sorted(row_data['Age']))
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12,6)) #열1개 컬럼3개 지정 size는 12*6인치
#fig는 전체 figure의 크기
#ax[0], ax[1], ax[2] 이렇게 접근.


#판다스는 matplotlib 없어도 시각화 함수를 자체적으로 가지고 있음. 다만 방법이 좀 다름.
ax[0].set_title('Survived-pie')
row_data['Survived'].value_counts().plot.pie(autopct='%.1f%%', ax=ax[0])
ax[0].set_ylabel('')

ax[1].set_title('Survived-bar')
row_data['Survived'].value_counts().plot.bar(ax=ax[1])

# ax[1].set_ylabel('명')

ax[2].set_title('Survived-seaborn')
sns.countplot(data=row_data, x='Survived', ax=ax[2])
```

```python
row_data['Survived'].value_counts().plot.pie(autopct='%.1f%%', explode=[0.1, 0])
```
