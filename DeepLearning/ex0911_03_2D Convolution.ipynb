```python
from tensorflow import keras
import cv2
import matplotlib.pyplot as plt

image_url = 'https://sipi.usc.edu/database/download.php?vol=misc&img=4.1.06'
filename = 'tree.tiff'
# image_path = keras.utils.get_file(filename, image_url)

# 이미지를 로드하고 크기를 조절합니다.
source = cv2.imread(filename)
image = cv2.resize(source, dsize=(80, 80), interpolation=cv2.INTER_AREA)

# 이미지를 표시합니다.
plt.figure(figsize=(10, 4))
plt.subplot(121), plt.imshow(image), plt.title('Original')

# 플롯을 표시합니다.

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

source = cv2.imread(filename)
image = cv2.resize(source, dsize=(80, 80), interpolation=cv2.INTER_AREA)

box_blur = np.array([[1, 1, 1],
                     [1, 1, 1],
                     [1, 1, 1]])/9

# 박스 블러 필터
dest = cv2.filter2D(image, -1, box_blur)

# 이미지 표시
plt.figure(figsize=(10, 4))
plt.subplot(121), plt.imshow(image), plt.title('Original')
plt.subplot(122), plt.imshow(dest), plt.title('Box Blur')

# 플롯을 표시합니다.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

source = cv2.imread(filename)
image = cv2.resize(source, dsize=(80, 80), interpolation=cv2.INTER_AREA)

edge_detect = np.array([[-1, -1, -1],
                     [-1, 8, -1],
                     [-1, -1, -1]])

# 엣지 디텍트 필터
dest = cv2.filter2D(image, -1, edge_detect)

# 이미지 표시
plt.figure(figsize=(10, 4))
plt.subplot(121), plt.imshow(image), plt.title('Original')
plt.subplot(122), plt.imshow(dest), plt.title('Edge_detect')

# 플롯을 표시합니다.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

source = cv2.imread(filename)
image = cv2.resize(source, dsize=(80, 80), interpolation=cv2.INTER_AREA)

sharpen = np.array([[0, -1, 0],
                     [-1, 5, -1],
                     [0, -1, 0]])

# 샤픈 필터
dest = cv2.filter2D(image, -1, sharpen)

# 이미지 표시
plt.figure(figsize=(10, 4))
plt.subplot(121), plt.imshow(image), plt.title('Original')
plt.subplot(122), plt.imshow(dest), plt.title('Sharpen')

# 플롯을 표시합니다.
plt.show()
```
