```python
class_names=['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```

```python
import numpy as np
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
```

```python
x_train = np.array([[0.2],[0.8]])
y_train = np.array([[0.3],[0.9]])

input_layer = Input((1))
output_layer = Dense(1, activation = None)(input_layer)
model = Model(input_layer, output_layer)

model.summary()
```

```python
layer = model.layers[1]
weights = layer.get_weights()

w_start = weights[0][0][0]
b_start = weights[1][0]
print(weights[0])
print(weights[1])
print(w_start, b_start)
```

```python
from matplotlib import pyplot as plt
from matplotlib import cm
```

```python
w = w_start
b = b_start

fig, ax = plt.subplots()

#여기부터는 그래프 세팅설정
ax.set_aspect('equal')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.axis([0,1,-1,1])
ax.scatter(x_train, y_train, c=y_train, cmap=cm.coolwarm)
x = np.linspace(0,1,51)
y = w * x + b
ax.plot(x,y)
plt.show()
```

```python
from tensorflow.keras.optimizers import Adam, SGD
model.compile(loss='mean_squared_error', optimizer=SGD(lr=0.05), metrics=['mse'])

model.fit(x_train, y_train, epochs=800)
```

```python
layer = model.layers[1]
weights = layer.get_weights()

w_start = weights[0][0][0]
b_start = weights[1][0]

w = w_start
b = b_start

fig, ax = plt.subplots()

#여기부터는 그래프 세팅설정
ax.set_aspect('equal')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.axis([0,1,-1,1])
ax.scatter(x_train, y_train, c=y_train, cmap=cm.coolwarm)
x = np.linspace(0,1,51)
y = w * x + b
ax.plot(x,y)
plt.show()
```
