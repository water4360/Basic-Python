```python
import pandas as pd
import numpy as np
data_loc = 'https://github.com/dongupak/DataML/raw/main/csv/'
df = pd.read_csv(data_loc + 'two_classes.csv')
df.tail(3)
```

```python
df_positive = df[df['y']==1]
df_negative = df[df['y']==0]

import matplotlib.pyplot as plt

plt.scatter(df_positive['x1'], df_positive['x2'], color='r')
plt.scatter(df_negative['x1'], df_negative['x2'], color='b')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
X = df[['x1', 'x2']].to_numpy()
y = df['y']
svm = Pipeline([
    ('scaler', StandardScaler()),
    ('linearSVC', LinearSVC(C=1, loss='hinge'))
])
svm.fit(X, y)
```

```python
#x1 속성값이 0보다 크면1, 0보다 같거나 작으면 0으로 레이블링
svm.predict([[0.12, 0.56], [-4, 40], [0, 40], [5,20]]) #결과 : [1, 0, 0, 1]
```
